```python
# Part 3 - Making new predictions
from tensorflow.keras.models import load_model
classifier = load_model('lion_tiger_vgg16.h5')
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image
```

```python
test_image = image.load_img('dataset/single_prediction/lion_or_tiger_3.jpg',
                            target_size = (224,224))
test_image = image.img_to_array(test_image)

test_image = np.expand_dims(test_image, axis = 0)

result = classifier.predict(test_image)
print(result)

# 0: cat, 1: dog, sigmoid có ngưỡng 0.5
if result[0][0] > result[0][1]:
  prediction = 'lion'
else:
  prediction = 'tiger'
print(prediction)
```

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img=mpimg.imread('dataset/single_prediction/lion_or_tiger_3.jpg')
imgplot = plt.imshow(img)
plt.show()
```
